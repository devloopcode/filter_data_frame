# 2. Create a column based on 1 condition: np.where()

```python
import numpy as np
```

```python
data['Price tier'] = price_desc
```

# 3. Create a conditional column from more than 2 choices: np.select()

```python
conditions = [
    data['Price_euros'] > 3_000,
    (data['Price_euros'] > 2_000) & (data['Price_euros'] <= 3_000),
    (data['Price_euros'] > 800) & (data['Price_euros'] <= 2_000),
    data['Price_euros'] <= 800
]
values = ['Too Expensive', 'Expensive', 'Affordable', 'Cheap']
```

```python
data['Price_Category'] = np.select(conditions, values, default='Unknown')
```

```python
data.value_counts('Price_Category')
```

```python
conditions = [
    data['Inches'] > 16,
    data['Inches'] > 14,
    data['Inches'] > 12,
    data['Inches'] < 12
]
values = ['Too Big', 'Big', 'Small', 'Too Small']
```

```python
data['Screen Size'] = np.select(conditions, values, default='Unknown')
```

```python
data
```

```python
data.value_counts('Screen Size')
```

```python
conditions = [
    data['Inches'] > 16,
    (data['Inches'] > 14) & (data['Inches'] <= 16),
    (data['Inches'] > 12) & (data['Inches'] <= 14),
    data['Inches'] <= 12
]
values = ['Too Big', 'Big', 'Small', 'Too Small']
```

```python
data['Screen Size'] = np.select(conditions, values, default='Unknown')
```

```python
data.value_counts('Screen Size')
```
